```python
from KID import KID
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = KID(kbT = .3*86.17,Qc=4e5)
tescarr = np.linspace(.01,1.,5)
dNqp = k.Nqp_0*1.1
```

```python
for tesc in tescarr:
    k.tesc = tesc
    k.plot_dAthetaresp(k.Nqp_0 + dNqp,tStop = .5e10,plot='dA')
plt.legend(tescarr)
```

```python
for tesc in tescarr:
    k.tesc = tesc
    k.plot_dAthetaresp(k.Nqp_0 + dNqp,tStop = .5e10,plot='theta')
plt.yscale('log')
plt.legend(tescarr)
```

```python
for tesc in tescarr:
    k.tesc = tesc
    k.plot_Nqpt(k.Nqp_0 + dNqp,plot_lin=False)
plt.legend(tescarr)
```
