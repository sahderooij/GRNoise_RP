```python
from KID import KID
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = KID()
```

```python
Qcarr = np.logspace(1,5,5)
for Qc in Qcarr:
    k.Qc = Qc
    dNqp = .6582*3e8/1000/k.D_0
    k.plot_dAthetaresp(k.Nqp_0 + dNqp,tStop = .5e10,plot='theta')
    plt.yscale('log')
plt.legend(Qcarr)
```
