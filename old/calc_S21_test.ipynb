```python
from calc_funcs import *
import numpy as np
import matplotlib.pyplot as plt

hw = 1.
D = 17.6
D0 = 17.6
kbT = D/8

s1,s2 = calc_cinduct(hw,D,kbT)

ak = .7
d = .06
lbd0 = .092
Qi = calc_Qi(s1,s2,ak,lbd0,d,D,D0,kbT)

Qc = 2*10**5
w0 = 5.

for dw in np.arange(-10**-3,10**-3,10**-5):
    S21 = calc_S21(Qi,Qc,w0,dw,w0)
    A = calc_A(S21,Qi,Qc)
    theta = calc_theta(S21,Qi,Qc)
    plt.figure(1)
    plt.plot(np.real(S21),np.imag(S21),'r.')
    plt.figure(2)
    plt.plot(dw,A,'r.')
    plt.figure(3)
    plt.plot(dw,theta,'r.')
```
